# Import data

```python
import pandas as pd
import numpy as np
df = pd.read_csv('data.csv')
df
```

# Data pre-processing

```python
df['Date_Time']= pd.to_datetime(df['Date_Time'])
df
```

```python
for col in df.columns[1:]:
    df[col]=df[col].replace(0,df[col].mean())
    df[col]=df[col].fillna(df[col].mean())
    
df
```

# Test Train Split

```python
from sklearn.model_selection import TimeSeriesSplit

tss = TimeSeriesSplit(n_splits = 3)
```

```python
df.set_index('Date_Time', inplace=True)
df.sort_index(inplace=True)
X = df.drop(labels=['DST index'], axis=1)
y = df['DST index']
```

```python
for train_index, test_index in tss.split(X):
    X_train_val, X_test = X.iloc[train_index, :], X.iloc[test_index,:]
    y_train_val, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
for train_index, val_index in tss.split(X_train_val):
    X_train, X_val = X_train_val.iloc[train_index, :], X_train_val.iloc[val_index,:]
    y_train, y_val = y_train_val.iloc[train_index], y_train_val.iloc[val_index]
```

# Model Fit for LSTM

```python
from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import *

from tensorflow.keras.callbacks import ModelCheckpoint

from tensorflow.keras.losses import MeanSquaredError

from tensorflow.keras.metrics import RootMeanSquaredError

from tensorflow.keras.optimizers import Adam

model1 = Sequential()

model1.add(InputLayer((9,1)))

model1.add(LSTM(64))

model1.add(Dense (8, 'relu'))

model1.add(Dense (1, 'linear'))

model1.summary()
```

```python
cp=ModelCheckpoint('model1/', save_best_only=True)

model1.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.00001), metrics=[RootMeanSquaredError()])
```

```python
from tensorflow.keras.optimizers import Adam

model1.compile(loss='mean_squared_error', optimizer=Adam(lr=0.01))

history = model1.fit(X_train, y_train, 
                     validation_data=(X_val, y_val), 
                     epochs=100, 
                     batch_size=32, 
                     callbacks=[cp])
```

```python
train_predictions=model1.predict(X_train).flatten() 
train_results=pd.DataFrame(data={'Train Predictions': train_predictions, 'Actuals':y_train}) 
train_results
```

```python
test_predictions=model1.predict(X_test).flatten() 
test_results=pd.DataFrame(data={'Test Predictions': test_predictions, 'Actuals':y_test}) 
test_results
```

# Performance Analysis

# Model Fit for SARIMAX

```python
train_data = pd.concat([X_train, y_train], axis=1)

val_data = pd.concat([X_val, y_val], axis=1)

test_data = pd.concat([X_test, y_test], axis=1)
train_data.index=pd.to_datetime(train_data.index)
train_data = train_data.asfreq('H')
train_endog = train_data['DST index']
train_data
test_data.index=pd.to_datetime(test_data.index)
test_data = test_data.asfreq('H')
test_endog = test_data['DST index']
test_data
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima.arima import auto_arima

model_1 = auto_arima(train_endog, seasonal=False, error_action='ignore', suppress_warnings=True, stepwise=True)
model = SARIMAX(train_endog, order=model_1.order, seasonal_order=(2, 1, 1, 35))
results = model.fit()
```

```python
train_pred = results.predict(start=train_endog.index[0], end=train_endog.index[-1], dynamic=False)
train_results_sarimax=pd.DataFrame(data={'Train Predictions': train_pred, 'Actuals':y_train}) 
train_results_sarimax
```
